```python
import networkx as nx
import matplotlib.pyplot as plt
import imageio
import os

# class ChordGraph:
#     def __init__(self):
#         self.G = nx.MultiDiGraph()
#         self.step_counter = 1


def midi_to_note_name(midi_number):
    if midi_number == 0:
        return "Rest"
    note_names = ['C', 'C#/Db', 'D', 'D#/Eb', 'E', 'F', 'F#/Gb', 'G', 'G#/Ab', 'A', 'A#/Bb', 'B']
    note_name = note_names[midi_number % 12]
    octave = midi_number // 12 - 1
    return note_name

class ChordGraph:
    def __init__(self):
        self.G = nx.MultiDiGraph()
        self.step_counter = 1
        self.chord_steps = {}  # Adapted to track notes/rests
        self.previous_chord_id = None

    def add_chord_progression(self, chord, beat_duration):
        if chord not in self.chord_steps:
            self.chord_steps[chord] = [self.step_counter]
            self.G.add_node(chord, chord=chord, step=[self.step_counter], duration=[beat_duration])
        else:
            self.chord_steps[chord].append(self.step_counter)
            self.G.nodes[chord]['step'].append(self.step_counter)
            self.G.nodes[chord]['duration'].append(beat_duration)

        # If there's a previous chord in the progression, connect it
        if self.previous_chord_id and self.previous_chord_id != chord:
            self.G.add_edge(self.previous_chord_id, chord)

        self.previous_chord_id = chord
        self.step_counter += beat_duration

    # Visualization and animation methods remain unchanged
    def process_melody_file(self, file_path):
        self.previous_chord_id = None  # Reset for each new file
        with open(file_path, 'r') as file:
            for line in file:
                note_midi, duration = map(int, line.strip().split())
                note_name = midi_to_note_name(note_midi)  # Convert MIDI number to note name
                # Treat rests specially to avoid merging all rests into a single node
                chord_id = note_name
                self.add_chord_progression(chord_id, duration)

    def print_cycles(self, full_loop=True, per_beat=False):
        cycles = list(nx.simple_cycles(self.G))
        print(f"Total number of cycles: {len(cycles)}")
        for i, cycle in enumerate(cycles, start=1):
            # Sort the cycle based on the first occurrence (step) of each node within the cycle
            sorted_cycle = sorted(cycle, key=lambda node: min(self.G.nodes[node]['step']))

            # Initialize a list to hold the formatted progression
            progression_list = []

            for node in sorted_cycle:
                chord_name = self.G.nodes[node]['chord']
                # Assume duration is stored as a list of durations for each node; take the first as the representative duration
                duration = self.G.nodes[node]['duration'][0] if 'duration' in self.G.nodes[node] else 1

                if per_beat:
                    # Repeat the chord name for each beat of its duration
                    chord_repetitions = [chord_name for _ in range(duration)]
                    progression_list.extend(chord_repetitions)
                else:
                    progression_list.append(chord_name)

            # Convert progression list to a string
            progression = " -> ".join(progression_list)

            if full_loop:
                # If full_loop is True, loop back to the start for clarity
                if not per_beat or (per_beat and duration > 1):
                    # Avoid adding the start again if per_beat is true and duration is 1, as it's already included
                    progression += " -> " + progression_list[0] if not per_beat else ''
            else:
                # Exclude the final node from the progression to avoid repeating the start
                if per_beat and duration > 1:
                    # If per_beat, exclude repeated instances at the end based on duration
                    progression = " -> ".join(progression_list[:-duration])

            print(f"Cycle {i}: {progression}")

    def visualize_graph(self):
        plt.figure(figsize=(15, 10))
        pos = nx.kamada_kawai_layout(self.G)
        nx.draw_networkx_nodes(self.G, pos, node_size=2500, node_color="skyblue", alpha=0.6)
        nx.draw_networkx_edges(self.G, pos, arrowstyle="->", arrowsize=20, edge_color="gray", alpha=0.6)
        labels = {node: f'{data["chord"]}\n(Step: {data["step"]}, Duration: {data["duration"]})' for node, data in self.G.nodes(data=True)}
        nx.draw_networkx_labels(self.G, pos, labels, font_size=10)
        plt.title("Chord Progression Graph", size=15)
        plt.axis('off')
        plt.show()

    def print_graph(self):
        for node, data in self.G.nodes(data=True):
            # Since we're now storing individual values, not lists, we access them directly
            print(f'{node} (Step: {data["step"]}, Duration: {data["duration"]})')
        print('\nTotal Nodes:', len(self.G.nodes()), '\nTotal Edges:', len(self.G.edges()))

    def visualize_progression(self, pos, highlight_nodes=None):
        plt.figure(figsize=(15, 10))
        nx.draw_networkx_nodes(self.G, pos, node_size=2000, node_color="skyblue", alpha=0.6)
        if highlight_nodes:
            nx.draw_networkx_nodes(self.G, pos, nodelist=highlight_nodes, node_size=2000, node_color="red", alpha=0.9)
        nx.draw_networkx_edges(self.G, pos, arrowstyle="->", arrowsize=20, edge_color="gray", alpha=0.6)
        labels = {node: node for node in self.G.nodes()}
        nx.draw_networkx_labels(self.G, pos, labels, font_size=12)
        plt.axis('off')
        # Move plt.close() to the caller function after plt.savefig() is called.

    def create_animation(self):
        pos = nx.kamada_kawai_layout(self.G)  # Calculate positions once for all nodes
        filenames = []
        currently_highlighted = []  # Keep track of nodes currently highlighted

        for current_step in range(1, self.step_counter):
            plt.figure(figsize=(15, 10))

            # Draw default nodes
            nx.draw_networkx_nodes(self.G, pos, node_size=2000, node_color="skyblue", alpha=0.6)
            nx.draw_networkx_edges(self.G, pos, arrowstyle="->", arrowsize=20, edge_color="gray", alpha=0.6)
            labels = {node: node for node in self.G.nodes()}
            nx.draw_networkx_labels(self.G, pos, labels, font_size=12)

            # Update the list of currently highlighted nodes
            for node, data in self.G.nodes(data=True):
                start_steps = data['step']
                durations = data['duration']
                for start, duration in zip(start_steps, durations):
                    if current_step >= start and current_step < start + duration:
                        if node not in currently_highlighted:
                            currently_highlighted.append(node)
                    elif node in currently_highlighted and current_step >= start + duration:
                        currently_highlighted.remove(node)

            # Highlight the currently highlighted nodes
            if currently_highlighted:
                nx.draw_networkx_nodes(self.G, pos, nodelist=currently_highlighted, node_size=2000, node_color="red", alpha=0.9)

            plt.axis('off')
            filename = f"frame_{current_step}.png"
            plt.savefig(filename)
            filenames.append(filename)
            plt.close()

        # Create the animated GIF
        with imageio.get_writer('chord_progression.mp4', mode='I', fps=10) as writer:
            for filename in filenames:
                image = imageio.imread(filename)
                writer.append_data(image)
                os.remove(filename)  # Cleanup

        print("Animation created: 'chord_progression.gif'")

    def process_file(self, file_path):
        self.previous_chord_id = None  # Reset for each file processing
        with open(file_path, 'r') as file:
            for line in file:
                parts = line.strip().split()
                chord = parts[0].split('/')[0]  # Get chord name, ignoring after '/'
                beat_duration = int(parts[-1])  # Duration is the last part
                self.add_chord_progression(chord, beat_duration)


chord_graph = ChordGraph()
chord_graph.process_file('finalized_chord.txt')  # Use the correct path to your melody file
chord_graph.visualize_graph()
chord_graph.print_cycles(full_loop=False, per_beat=True)  #
# chord_graph.create_animation()  # Uncomment to create and save the animation


melody_graph = ChordGraph()
melody_graph.process_melody_file('melody.txt')  # Use the correct path to your melody file
melody_graph.visualize_graph()
melody_graph.print_cycles(full_loop=False, per_beat=False)  #
melody_graph.create_animation()  # Uncomment to create and save the animation




```
